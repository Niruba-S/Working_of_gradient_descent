```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

**VALUES OF X**

```python
X=np.arange(10)
X
```

**LETS DEFINE Y WHICH IS THE FUNCTION OF X**

```python
Y=(X-5)**2
Y
```

**GIVEN A FUNCTION f(X),FINDING VALUE OF X THAT MINIMIZES F**

# VISUALISATION

```python
plt.plot(X,Y)
plt.style.use("seaborn")
plt.ylabel("F(X)")
plt.xlabel("X")
```

**DEFINING INITIAL VALUE AND STEP SIZE**

```python
x=0
lr=0.1
error=[]
plt.plot(X,Y)
for i in range(50): 
 grad=2*(x-5)
 x=x-lr*grad
 y=(x-5)**2
 error.append(y)
 print(x)
 plt.scatter(x,y)
```

**PLOTTING ERROR**

```python
plt.plot(error)
plt.style.use("seaborn")
plt.xlabel("ERROR")
```
